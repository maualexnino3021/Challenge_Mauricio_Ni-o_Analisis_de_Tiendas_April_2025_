### Importación de datos



```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de estilo
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
# URLs de los datasets
urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
]

# Función para estandarizar nombres de columnas
def estandarizar_columnas(df):
    mapeo_columnas = {
        'categoria': ['categoría', 'category', 'categoria', 'Categoria', 'CATEGORIA', 'tipo', 'type'],
        'producto': ['producto', 'product', 'Producto', 'PRODUCTO', 'item', 'articulo'],
        'precio': ['precio', 'price', 'Precio', 'PRECIO', 'valor', 'costo'],
        'calificación': ['calificación', 'rating', 'Calificación', 'CALIFICACIÓN', 'puntuacion', 'score'],
        'envío': ['envío', 'shipping', 'Envío', 'ENVÍO', 'tiempo_entrega', 'dias_envio', 'delivery']
    }

    for estandar, alternativas in mapeo_columnas.items():
        for col in df.columns:
            if col.lower() in [alt.lower() for alt in alternativas]:
                df.rename(columns={col: estandar}, inplace=True)
    return df
```

```python
# Cargar datos y estandarizar nombres
tiendas = []
for i, url in enumerate(urls, 1):
    try:
        df = pd.read_csv(url)
        print(f"\nColumnas originales Tienda {i}: {df.columns.tolist()}")
        df = estandarizar_columnas(df)
        df['tienda'] = f'Tienda {i}'
        tiendas.append(df)
        display(df.head())
    except Exception as e:
        print(f"Error al cargar Tienda {i}: {str(e)}")

# Concatenar todos los datos
df_completo = pd.concat(tiendas, ignore_index=True)

# Verificación de columnas estandarizadas
print("\nColumnas finales del DataFrame completo:", df_completo.columns.tolist())
```

#1. Análisis de facturación



```python
print("\n1. Análisis de facturación por tienda")
if 'precio' in df_completo.columns:
    facturacion = df_completo.groupby('tienda')['precio'].sum().sort_values(ascending=False)
    print(facturacion)

    facturacion.plot(kind='bar', color='skyblue')
    plt.title('Facturación Total por Tienda')
    plt.xlabel('Tienda')
    plt.ylabel('Facturación Total')
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
else:
    facturacion = pd.Series(dtype=float)
    print("⚠️ No se encontró la columna 'precio'")

```

# 2. Ventas por categoría

```python
print("\n2. Ventas por categoría")
if 'categoria' in df_completo.columns:
    df_completo['categoria'] = df_completo['categoria'].str.lower().str.strip()
    ventas_categoria = df_completo.groupby(['tienda', 'categoria']).size().unstack().fillna(0)

    ventas_categoria.plot(kind='bar', stacked=True, cmap='viridis')
    plt.title('Ventas por Categoría en cada Tienda')
    plt.xlabel('Tienda')
    plt.ylabel('Cantidad de Ventas')
    plt.xticks(rotation=45)
    plt.legend(title='Categoría', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()
else:
    print("⚠️ No se encontró la columna 'categoria'")

```

# 3. Calificación promedio de la tienda


```python
print("\n3. Calificación promedio por tienda")
if 'calificación' in df_completo.columns:
    rating_promedio = df_completo.groupby('tienda')['calificación'].mean().sort_values(ascending=False)
    print(rating_promedio)

    rating_promedio.plot(kind='pie', autopct='%1.1f%%',
                         colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen'])
    plt.title('Calificación Promedio por Tienda')
    plt.ylabel('')
    plt.show()
else:
    rating_promedio = pd.Series(dtype=float)
    print("⚠️ No se encontró la columna 'calificación'")

```

# 4. Productos más y menos vendidos

```python
print("\n4. Productos más y menos vendidos")
if 'producto' in df_completo.columns:
    top_productos = df_completo['producto'].value_counts().head(5)
    bottom_productos = df_completo['producto'].value_counts().tail(5)

    print("Top 5 más vendidos:\n", top_productos)
    print("\nBottom 5 menos vendidos:\n", bottom_productos)

    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    top_productos.plot(kind='barh', color='green')
    plt.title('Top 5 Productos Más Vendidos')

    plt.subplot(1, 2, 2)
    bottom_productos.plot(kind='barh', color='red')
    plt.title('Bottom 5 Productos Menos Vendidos')

    plt.tight_layout()
    plt.show()
else:
    print("⚠️ No se encontró la columna 'producto'")
```

# 5. Envío promedio por tienda

```python
print("\n5. Tiempo de envío promedio por tienda")

columnas_envio = [col for col in df_completo.columns
                  if any(k in col.lower() for k in ['env', 'shipping', 'entrega', 'delivery', 'tiempo'])]

if columnas_envio:
    col_envio = columnas_envio[0]
    print(f"Usando columna '{col_envio}'")

    df_completo[col_envio] = pd.to_numeric(df_completo[col_envio], errors='coerce').fillna(0)
    envio_promedio = df_completo.groupby('tienda')[col_envio].mean().sort_values()

    print(envio_promedio)

    envio_promedio.plot(kind='line', marker='o', color='purple')
    plt.title(f'Tiempo de Envío Promedio por Tienda\n(Columna: {col_envio})')
    plt.xlabel('Tienda')
    plt.ylabel('Días de Envío Promedio')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()
else:
    envio_promedio = pd.Series(dtype=float)
    print("⚠️ No se encontró ninguna columna relacionada con envíos")
```

```python
print("\nRecomendación final basada en el análisis:")

def calcular_puntaje(tienda):
    puntaje = 0
    if not facturacion.empty:
        puntaje += facturacion.get(tienda, 0) / facturacion.max() * 40
    if not rating_promedio.empty:
        puntaje += rating_promedio.get(tienda, 0) / rating_promedio.max() * 30
    if not envio_promedio.empty:
        puntaje += (1 - envio_promedio.get(tienda, 0) / envio_promedio.max()) * 30
    return puntaje

if len(tiendas) > 0:
    puntajes = {tienda: calcular_puntaje(tienda) for tienda in df_completo['tienda'].unique()}
    mejor_tienda = max(puntajes.items(), key=lambda x: x[1])[0]

    print(f"""
    Basándonos en los datos analizados, la mejor tienda para que el Sr. João venda sus productos sería:

    1. Facturación Total: {facturacion.idxmax() if not facturacion.empty else "No disponible"}
    2. Calificación Promedio: {rating_promedio.idxmax() if not rating_promedio.empty else "No disponible"}
    3. Tiempo de Envío: {envio_promedio.idxmin() if not envio_promedio.empty else "No disponible"}

    Considerando un análisis ponderado de estos factores (Facturación 40%, Calificación 30%, Envío 30%),
    la recomendación es: {mejor_tienda}
    """)

    comparacion = pd.DataFrame({
        'Facturación': facturacion if not facturacion.empty else None,
        'Calificación': rating_promedio if not rating_promedio.empty else None,
        'Envío Promedio': envio_promedio if not envio_promedio.empty else None,
        'Puntaje Total': pd.Series(puntajes)
    })
    display(comparacion.sort_values('Puntaje Total', ascending=False))
else:
    print("⚠️ No se pudo cargar ningún dataset para realizar el análisis.")
```
